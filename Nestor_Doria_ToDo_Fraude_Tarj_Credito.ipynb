## 3. Verificar si hay variables categóricas y codificarlas si es necesario.

La variable categórica sería la de "Class", debido a que es la unica que cuenta con unos y ceros.

```python
from sklearn.preprocessing import OneHotEncoder
def codificar(df,col):
  # Ingresar el dataframe df y una lista de las columnas categóricas col.
  df = dataset.copy()
  codificador = OneHotEncoder()
  codificacion = codificador.fit_transform(df[col]).toarray()
  codificacion = pd.DataFrame(codificacion, columns=np.concatenate(codificador.categories_))
  df.drop(col, axis=1, inplace=True)
  df = pd.concat([df,codificacion],axis=1) 
  return df
```

En la columna 0 es fraude, y en la columna 1 no es fraude

```python
data_encod = codificar(dataset,['Class'])
data_encod.head()
```

## 4. Crear los conjuntos de entrenamiento (train) y testeo (test).

```python
import pandas as pd
from sklearn.model_selection import train_test_split
X = dataset.drop(columns='Class')
y = dataset["Class"] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## 5. Implementar Regresión Logística, con las métricas correspondientes y la matriz de confusión. Analizar los resultados.

```python
from sklearn.linear_model import LogisticRegression
classifier_lr = LogisticRegression()
classifier_lr.fit(X_test, y_test)
```

```python
y_pred = classifier_lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
```

```python
y_pred1 = classifier_lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
acc = accuracy_score(y_test, y_pred1)
prec = precision_score(y_test, y_pred1)
rec = recall_score(y_test, y_pred1)
f1 = f1_score(y_test, y_pred1)

results = pd.DataFrame([['lr', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
```

```python
results
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred1)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
